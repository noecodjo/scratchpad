```python
import pefile
```

```python
p = pefile.PE("layer2.exe")
print p.dump_info()
```

```python
# dump the interesting resources
for entry in p.DIRECTORY_ENTRY_RESOURCE.entries:
    if str(entry.name) == 'ACCELORATOR':
        for e in entry.directory.entries:
            if e.id == 0x8A or e.id == 0x8C:
                if isinstance(e, pefile.ResourceDirEntryData):
                    # if you get errors here, inspect the underlying structure because
                    # it changed
                    offset = e.directory.entries[0].data.struct.OffsetToData
                    size = e.directory.entries[0].data.struct.Size
                    
                    # extract data
                    payload = p.get_memory_mapped_image()[offset:offset+size]
                    
                    # write to file
                    with open("extracted_{}".format(e.id), 'wb') as f:
                        f.write(payload)
```
